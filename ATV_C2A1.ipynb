<strong>Etapa de Importações</strong>

```python
'''libs'''
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

'''DataFrames'''
data = pd.read_csv("./Placement_Data_Full_Class.csv")
df_original = pd.DataFrame(data)
df_original
```

<strong>Etapa de Transformações:
</strong>

```python
'''cópia do Data Frame'''
df_no_salary = df_original.set_index('sl_no') #transformando a coluna de ID no index do DF
df_no_salary.pop('salary')
df_no_salary
```

<strong>Etapa de Análise Exploratória de Dados</strong>

```python
df_original.info() #verificando valores nulos e tipos de dados
df_original.describe() #aplicando cálculos prévios para entender as colunas numéricas
```

```python
 #verificando de ocorrências de acordo com as colunas categóricas (não numéricas)
cols_categoricas = ['gender','hsc_s','degree_t','workex','specialisation','status']
for col in cols_categoricas:
    sns.countplot(data=df_original, x=col)
    plt.xticks(rotation=45)
    plt.show()
```
